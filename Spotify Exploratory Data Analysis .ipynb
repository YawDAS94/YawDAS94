```python
# Extract the year from 'release_date' and set it as a new column
df_tracks['release_year'] = pd.to_datetime(df_tracks['release_date'], errors='coerce').dt.year

# Drop rows where the 'release_year' could not be converted (NaT values)
df_tracks.dropna(subset=['release_year'], inplace=True)

# Convert 'release_year' to integer
df_tracks['release_year'] = df_tracks['release_year'].astype(int)


# Set 'release_year' as the index
df_tracks.set_index('release_year', inplace=True)

# Display the first few rows to confirm the changes
df_tracks.head()
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round (x/1000)) #turn ms oto seconds
df_tracks.drop("duration_ms", inplace=True,axis=1)
```

```python
df_tracks.duration.head(15)
```

```python
# Drop non-numeric columns and those you don't need
numeric_df = df_tracks.select_dtypes(include=[float, int]).drop(["key", "mode", "explicit"], axis=1)

# Compute the correlation matrix
corr_df = numeric_df.corr(method="pearson")


# Plot the heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='plasma', linewidths=1, linecolor="black")
plt.show()
```

```python
#energy and acousticness are negatively correlated, energy is high, acousticness is low
#energy and loud are positively correlated
```

```python
sample_df = df_tracks.sample(int(0.004 *len(df_tracks))) #0.4% sample size
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness",x = "energy", color = "g").set(title = "Loudness vs. Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity",x = "acousticness", color = "c").set(title = "Popularity vs. Acousticness Correlation")
#relatively negatively correlated
```

```python
# Ensure the index is correctly set to the release year
df_tracks.index = pd.to_datetime(df_tracks.index, format='%Y').year

# Group by year and count the number of songs
songs_per_year = df_tracks.groupby(df_tracks.index).size()

# Filter the data to include only years since 1922
songs_per_year = songs_per_year[songs_per_year.index >= 1922]

# Plot the number of songs per year
plt.figure(figsize=(12, 6))
plt.plot(songs_per_year.index, songs_per_year.values, marker='o', linestyle='-')
plt.title('Number of Songs Released Each Year Since 1922')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```
