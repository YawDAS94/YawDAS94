<h1><b> Logistic Regression </b></h1>

```python
from sklearn.model_selection import train_test_split
```

```python
x = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .33, random_state=42)
```

```python
#traina nd fit a logistic regression model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train, y_train)
```

```python
predictions = logmodel.predict(x_test)
```

#make a classification report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

<h2><b>Feature Engineering</b></h2>

<h2><b> Why I am Feature Engineering with interaction terms</b></h2>
<b> I am making a new variable called time age interaction. The purpose of me making this variable is to capture the combined effect of two variables (age and time spent on site) to see if the new combined variable jointly influences the target varaible (clicked on ad) in a way that's not capturred by each variable independently

By making an interaction term, I am allowing the model to consider the possibiltiy that the impact of one variable on the target variable changes depending on the level with the other. So maybe time spent matters depeneding on how old the user is. Or maybe the age only matters when considering it with time spent. With the interaction varialbe, I can now see how time spent affects across different age groups.

For example:
If I have two users:
User A: Age 25, daily time spent  = 100
User B: Age 60, daily time spent = 100

without the interaction term, the model might predict the same likelihood of clicking on an ad for both users, if all variables are the same. However, maybe younger users like User A might be more likely to click on ads when they spend a lot of time on the site, while older users (like User b0 might be less influenced by time spent. Maybe the relationship is non linear, so interaction terms are user.

I think the only concern I may have about feature engineering is overfitting, where adding interaction terms increases complexity of the model which may lead to overfitting with this limited data. I will cross validate afterwards to valdiate the model's performance. 

</b> 


```python
ad_data['Time_Age_Interaction'] = ad_data['Daily Time Spent on Site'] * ad_data['Age']
```

```python
ad_data.head()
```

```python
x2 = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male','Time_Age_Interaction' ]]
y2 = ad_data['Clicked on Ad']
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2,y2, test_size = .33, random_state=42)
```

```python
logmodel2 = LogisticRegression()
```

```python
logmodel2.fit(x2_train, y2_train)
```

```python
predictions2 = logmodel2.predict(x2_test)
```

```python
print(classification_report(y2_test,predictions2))
```
